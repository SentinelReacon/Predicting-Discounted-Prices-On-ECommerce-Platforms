```python
import pandas as pd
```

```python
from sklearn.impute import SimpleImputer
import numpy as np
import seaborn as sns
```

```python
df_again.replace({'maincateg':{'Men':0,'Women':1}},inplace=True)
df_again.replace({'platform':{'Flipkart':0,'Amazon':1}},inplace=True)
```

```python
#Instantiate SimpleImputer 
si=SimpleImputer(missing_values = np.nan, strategy="median")
si.fit(df_again[['norating1', 'star_5f', 'noreviews1', 'star_4f', 'maincateg', 'star_3f']])
  
#Filling missing data with median
df_again[['norating1', 'star_5f', 'noreviews1', 'star_4f', 'maincateg', 'star_3f']] = si.transform(df_again[['norating1', 'star_5f', 'noreviews1', 'star_4f', 'maincateg', 'star_3f']])
```

```python
sns.pairplot(data=df_again, diag_kind='kde')
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (16,10))
ax=sns.heatmap(df_again.corr(), cmap='Blues', annot=True)
plt.show()
```

```python
df_again.dtypes
```

```python
df_again.drop(['title','Offer %','star_5f','star_4f','star_3f','star_2f','star_1f','maincateg'], axis=1, inplace=True)
```

```python
from sklearn.model_selection import train_test_split
xx = df_again[['id','Rating','platform','actprice1','norating1','noreviews1','fulfilled1']]
yy = df_again['price1']

xx_train, xx_test, yy_train, yy_test=train_test_split(xx,yy, test_size=0.3, random_state=200)
```

```python
#Scaling numeric features using sklearn StandardScalar
numeric=['id', 'Rating', 'actprice1','norating1','noreviews1']
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
xx_train[numeric]=sc.fit_transform(xx_train[numeric])
xx_test[numeric]=sc.transform(xx_test[numeric])
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import xgboost
xgb_again = xgboost.XGBRegressor()
```

```python
xgb_again.fit(xx_train, yy_train)
```

```python
yy_pred = xgb_again.predict(xx_test)
r2_score(yy_test, yy_pred), mean_absolute_error(yy_test, yy_pred), np.sqrt(mean_squared_error(yy_test, yy_pred))
```

```python
xgb_again.score(xx_test, yy_test)
```

```python
from sklearn.compose import TransformedTargetRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import QuantileTransformer

param_grid = {'n_estimators': [100, 80, 60, 55, 51, 45],  
              'max_depth': [7, 8],
              'reg_lambda' :[0.26, 0.25, 0.2]
             }
                
grid = GridSearchCV(xgboost.XGBRegressor(), param_grid, refit = True, verbose = 3, n_jobs=-1) #
regr_trans = TransformedTargetRegressor(regressor=grid, transformer=QuantileTransformer(output_distribution='normal'))

# fitting the model for grid search 
grid_result=regr_trans.fit(xx_train, yy_train)
best_params=grid_result.regressor_.best_params_
print(best_params)

#using best params to create and fit model
best_model = xgboost.XGBRegressor(max_depth=best_params["max_depth"], n_estimators=best_params["n_estimators"], reg_lambda=best_params["reg_lambda"])
regr_trans = TransformedTargetRegressor(regressor=best_model, transformer=QuantileTransformer(output_distribution='normal'))
regr_trans.fit(xx_train, yy_train)
yhat = regr_trans.predict(xx_test)

#evaluate metrics
r2_score(yy_test, yhat), mean_absolute_error(yy_test, yhat), np.sqrt(mean_squared_error(yy_test, yhat))
```
